```python
#@markdown # Connecting to ngrok without token... (Recommended)

!pip install -r requirements.txt
!pip install spandrel kornia

# نصب ngrok
!wget -q -O ngrok.zip https://bin.equinox.io/c/bNyj1mQVY4c/ngrok-v3-stable-linux-amd64.zip
!unzip -o ngrok.zip
!chmod +x ./ngrok  # تنظیم مجوز اجرایی برای ngrok


from IPython.display import clear_output

# پاک کردن خروجی ترمینال
clear_output()

import subprocess
import threading
import time
import socket
import requests

def iframe_thread(port):
    # صبر کردن برای شروع سرور محلی
    while True:
        time.sleep(0.5)
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        if result == 0:
            break
        sock.close()

    print("\nComfyUI finished loading, trying to launch ngrok...\n")

    # اجرای ngrok بدون احراز هویت
    subprocess.Popen(["./ngrok", "http", str(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    # صبر کردن برای شروع ngrok و دریافت لینک عمومی
    ngrok_url = None
    for _ in range(20):  # تلاش به مدت 10 ثانیه
        try:
            response = requests.get("http://127.0.0.1:4040/api/tunnels")
            tunnels = response.json().get("tunnels", [])
            for tunnel in tunnels:
                if tunnel.get("proto") == "https":
                    ngrok_url = tunnel.get("public_url")
                    break
            if ngrok_url:
                break
        except requests.exceptions.RequestException:
            pass
        time.sleep(0.5)

    if ngrok_url:
        print("This is the URL to access ComfyUI:", ngrok_url)
    else:
        print("Failed to retrieve ngrok URL.")

# اجرای ترد ngrok
threading.Thread(target=iframe_thread, daemon=True, args=(8188,)).start()

# اجرای فایل اصلی
!python main.py --dont-print-server
```

```python
#@markdown # Connecting to ngrok with token... (Recommended)


!pip install -r requirements.txt
!pip install spandrel kornia



# نصب ngrok
!wget -q -O ngrok.zip https://bin.equinox.io/c/bNyj1mQVY4c/ngrok-v3-stable-linux-amd64.zip
!unzip -o ngrok.zip
!chmod +x ./ngrok  # تنظیم مجوز اجرایی برای ngrok

from IPython.display import clear_output

# پاک کردن خروجی ترمینال
clear_output()

import subprocess
import threading
import time
import socket
import requests

def iframe_thread(port, ngrok_token):
    # صبر کردن برای شروع سرور محلی
    while True:
        time.sleep(0.5)
        sock = socket.socket(socket.AF_INET, socket.SOCK_STREAM)
        result = sock.connect_ex(('127.0.0.1', port))
        if result == 0:
            break
        sock.close()

    print("\nComfyUI finished loading, trying to launch ngrok...\n")

    # احراز هویت ngrok
    auth_command = ["./ngrok", "config", "add-authtoken", ngrok_token]
    subprocess.run(auth_command, check=True)

    # اجرای ngrok
    subprocess.Popen(["./ngrok", "http", str(port)], stdout=subprocess.PIPE, stderr=subprocess.PIPE)

    # صبر کردن برای شروع ngrok و دریافت لینک عمومی
    ngrok_url = None
    for _ in range(20):  # تلاش به مدت 10 ثانیه
        try:
            response = requests.get("http://127.0.0.1:4040/api/tunnels")
            tunnels = response.json().get("tunnels", [])
            for tunnel in tunnels:
                if tunnel.get("proto") == "https":
                    ngrok_url = tunnel.get("public_url")
                    break
            if ngrok_url:
                break
        except requests.exceptions.RequestException:
            pass
        time.sleep(0.5)

    if ngrok_url:
        print("This is the URL to access ComfyUI:", ngrok_url)
    else:
        print("Failed to retrieve ngrok URL.")

# وارد کردن توکن ngrok
ngrok_token = ""  # @param {type:"string"}

# اجرای ترد ngrok
threading.Thread(target=iframe_thread, daemon=True, args=(8188, ngrok_token)).start()

# اجرای فایل اصلی
!python main.py --dont-print-server
```
